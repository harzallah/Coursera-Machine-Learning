```python
# %load ../../standard_import.txt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import axes3d
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model

pd.set_option('display.notebook_repr_html', False)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 150)
pd.set_option('display.max_seq_items', None)
 
#%config InlineBackend.figure_formats = {'pdf',}
%matplotlib inline  

import seaborn as sns
sns.set_context('notebook')
sns.set_style('white')
```

```python
X = np.array([30, 50, 80, 120, 150])
X = X.reshape(-1,1)
y = np.array([60, 80, 95, 105, 110])

plt.scatter(X, y, s=30, c='r', marker='x', linewidths=2)
plt.xlim(0,151)
plt.ylabel('House Price in K TND')
plt.xlabel('Surface in m2');
```

```python
xx = np.arange(20,156, 5)
plt.scatter(X, y, s=30, c='r', marker='x', linewidths=1)

regr = LinearRegression()

regr.fit(X, y.ravel())
plt.plot(xx, regr.intercept_+regr.coef_*xx, label='Linear regression 1')

plt.xlim(0,155)
plt.ylabel('House Price in K TND')
plt.xlabel('Surface in m2');
plt.legend(loc=4);
```

```python
xx = np.arange(20,156, 5)

poly = PolynomialFeatures(2)
Xpoly = poly.fit_transform(X)
print 'New number of features :', Xpoly.shape[1]
plt.scatter(X, y, s=30, c='r', marker='x', linewidths=1)

# print Xpoly
regr = LinearRegression()
regr.fit(Xpoly, y.ravel())

theta = regr.coef_.reshape(-1, 1)
xxpoly = poly.fit_transform(xx.reshape(-1, 1)) 
yypoly = regr.intercept_+xxpoly.dot(theta)

plt.plot(xx, yypoly, label='Linear regression 2')

plt.xlim(0,155)
plt.ylabel('House Price in K TND')
plt.xlabel('Surface in m2');
plt.legend(loc=4);
```

```python
xx = np.arange(20,156, 5)

poly = PolynomialFeatures(6)
Xpoly = poly.fit_transform(X)
print 'New number of features :', Xpoly.shape[1]
plt.scatter(X, y, s=30, c='r', marker='x', linewidths=1)

# print Xpoly
regr = LinearRegression()
regr.fit(Xpoly, y.ravel())

theta = regr.coef_.reshape(-1, 1)
xxpoly = poly.fit_transform(xx.reshape(-1, 1)) 
yypoly = regr.intercept_+xxpoly.dot(theta)

plt.plot(xx, yypoly, label='Linear regression 6')

plt.xlim(0,155)
plt.ylabel('House Price in K TND')
plt.xlabel('Surface in m2');
plt.legend(loc=4);
```

```python
xx = np.arange(20,156, 5)

poly = PolynomialFeatures(10)
Xpoly = poly.fit_transform(X)
print 'New number of features :', Xpoly.shape[1]
plt.scatter(X, y, s=30, c='r', marker='x', linewidths=1)

regr = LinearRegression()
regr.fit(Xpoly, y.ravel())

theta = regr.coef_.reshape(-1, 1)
xxpoly = poly.fit_transform(xx.reshape(-1, 1)) 
yypoly = regr.intercept_+xxpoly.dot(theta)

plt.plot(xx, yypoly, label='Linear regression 10')

plt.xlim(0,155)
plt.ylabel('House Price in K TND')
plt.xlabel('Surface in m2');
plt.legend(loc=4);
```

```python
xx = np.arange(20,156, 5)

poly = PolynomialFeatures(10)
Xpoly = poly.fit_transform(X)
print 'New number of features :', Xpoly.shape[1]
plt.scatter(X, y, s=30, c='r', marker='x', linewidths=1)

regr = linear_model.Lasso(alpha = 0.1)
regr.fit(Xpoly, y.ravel())

theta = regr.coef_.reshape(-1, 1)
xxpoly = poly.fit_transform(xx.reshape(-1, 1)) 
yypoly = regr.intercept_+xxpoly.dot(theta)

plt.plot(xx, yypoly, label='Linear regression 10 + Regularization')

plt.xlim(0,155)
plt.ylabel('House Price in 1000 TND')
plt.xlabel('Surface in m2');
plt.legend(loc=4);
```
